```python
import music21 as m21

# Score main container
score = m21.stream.Score()

# Parts (instruments)
part = m21.stream.Part()
bass_part = m21.stream.Part()


voice1 = m21.stream.Voice()
voice2 = m21.stream.Voice()
bass_voice = m21.stream.Voice()

notes = ['C4', 'D4', 'E4', 'F4', 'G4' , 'A4', 'B4', 'C5']
for note in notes:
    melody_note = m21.note.Note(note);
    harmony_note = melody_note.transpose(-8)
    voice1.append(melody_note)
    voice2.append(harmony_note)

    bass_note = m21.note.Note(note);
    bass_note.octave -= 2
    bass_voice.append(bass_note)

# Create a measure and insert both voices at offset 0
measure = m21.stream.Measure()
measure.insert(0, voice1)
measure.insert(0, voice2)

bass_measure = m21.stream.Measure()
bass_measure.insert(0, bass_voice)
bass_part.append(bass_measure)

part.append(measure)
score.insert(0, part)
score.insert(0, bass_part)
score.show('midi')
```

```python
c_major = m21.chord.Chord('C5 E4 G3')
print("pitches", c_major.pitches)
print("commonName", c_major.commonName)
print(".quarterLength", c_major.quarterLength)
print(".duration.type", c_major.duration.type)
print(".duration.dots", c_major.duration.dots)
print(".isMajorTriad()", c_major.isMajorTriad())
print(".inversion()", c_major.inversion())
print(".isTriad()", c_major.isTriad())
print(".root", c_major.root())
print(".bass", c_major.bass())
print(".third", c_major.third)
print(".fifth", c_major.fifth)
print(".seventh", c_major.seventh)
c_major.show('text')
c_major.show()
c_major.show('midi')
```

```python
from music21 import key, scale, chord

k = key.Key('A')
s = k.getScale()          # C major scale

degrees = [1, 3, 5, 6]
pitches = [s.pitchFromDegree(d) for d in degrees]

c_major = chord.Chord(pitches)
print(c_major)
print(c_major.pitchNames)
c_major.show()
c_major.show('midi')
```
